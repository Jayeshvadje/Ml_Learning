```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False)
encode_data=ohe.fit_transform(df[["Department"]])
```

```python
encode_data
```

```python
encoded_df=pd.DataFrame(encode_data,columns=ohe.get_feature_names_out(["Department"]))
```

```python
encoded_df
```

```python
seniority_map={"Junior":3,"Mid":2,"Senior":1}
```

```python
df["Seniority_Level"]= df["Seniority_Level"].map(seniority_map)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Remote_Work"]=le.fit_transform(df["Remote_Work"])
```

```python
print(f"Classes found: {le.classes_}")
```

```python
df_final = pd.concat([df.drop('Department', axis=1), encoded_df], axis=1)
```
