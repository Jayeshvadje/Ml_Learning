```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df.drop("customer_id",axis=1,inplace=True)
df.drop("leak_feature",axis=1,inplace=True)

```

```python
x=df.drop(columns = "target",axis=1)
y=df["target"]
```

```python
x
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
best_score = 0
best_model = None
best_k = None

for K in [1,3,5,7,9,11]:

    knn = KNeighborsClassifier(n_neighbors=K)
    knn.fit(X_train, y_train)

    prediction = knn.predict(X_test)

    score = accuracy_score(y_test, prediction)

    print(K, "=", score)

    if score > best_score:
        best_score = score
        best_model = knn
        best_k = K
```

```python
print("Best K:", best_k)

train_score = best_model.score(X_train, y_train)
test_score = best_model.score(X_test, y_test)

print("Train:", train_score)
print("Test:", test_score)
```

```python
from sklearn.model_selection import cross_val_score

# 5-fold cross-validation
scores = cross_val_score(knn, x, y, cv=5)
print(f"Average CV Score: {scores.mean():.4f}")
```

```python
#main method to use this

from sklearn.model_selection import GridSearchCV

paramgrid = {"n_neighbors": np.arange(1,31)}
knn_gscv = GridSearchCV(KNeighborsClassifier(), param_grid=paramgrid, cv=5)
knn_gscv.fit(x,y)
print(f"Optimal K is: {knn_gscv.best_params_['n_neighbors']}")
```
