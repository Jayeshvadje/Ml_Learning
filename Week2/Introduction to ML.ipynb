1. Machine Learning Basics

    What is Machine Learning? A subset of Artificial Intelligence (AI) that enables computers to "learn" patterns from data and make decisions without being explicitly programmed for every specific task.

    Why Machine Learning is used: It is used to automate complex decision-making, identify hidden patterns in massive datasets, and make predictions where traditional "if-then" rules would be too complex or impossible to maintain.

    Real-life ML examples: Recommendation engines (Netflix/Amazon), Email spam filters, credit card fraud detection, and predictive maintenance in industrial IoT.

2. Types of Machine Learning

    Supervised Learning: The model is trained on "labeled" data (data with a known answer). It learns the mapping from input to output.

    Unsupervised Learning: The model works with "unlabeled" data. It looks for structure or patterns in the data itself without knowing the correct answers beforehand.

    Reinforcement Learning: An agent learns to make decisions by performing actions in an environment to maximize a "reward." It learns through trial and error.

3. Supervised Learning

    Labeled data: Datasets where every input (X) has a corresponding, known output (Y or "ground truth").

    Input features and target variable: Features are the independent variables (input data); the Target Variable is the dependent variable (what you want to predict).

    Training and testing data: The Training set is used to teach the model; the Test set is "hidden" data used to evaluate how well the model generalizes to new information.

4. Unsupervised Learning

    Unlabeled data: Data that has no associated tags or correct outputs. The computer must infer the structure.

    Clustering: Grouping similar data points together (e.g., segmenting customers based on purchasing behavior).

    Dimensionality reduction: The process of reducing the number of variables (features) in a dataset while keeping the most important information (e.g., PCA).

5. Regression vs Classification

    Regression: Used when the output is a continuous numeric value (e.g., predicting house prices, stock values, or temperature).

    Classification: Used when the output is a category or label (e.g., "Yes/No," "Spam/Not Spam," "Cat/Dog").

    Difference: Regression predicts "How much," while Classification predicts "Which class."

6. Common ML Workflow

    Data collection: Gathering raw data from databases, APIs, or files.

    Data cleaning: Handling missing values, removing duplicates, and fixing inconsistencies.

    Feature selection: Choosing the most relevant variables to improve model performance and speed.

    Model training: Feeding the prepared data into an algorithm to learn patterns.

    Model evaluation: Testing the model against new data to measure how well it performs.

7. Basic ML Terms

    Features: The inputs used for predictions (e.g., Age, Fare).

    Labels: The outcome variable the model tries to predict (e.g., "Survived").

    Predictions: The output generated by the model for new, unseen data.

    Training set: The portion of data used to "build" the model.

    Test set: The portion of data used to "validate" the model.

    Overfitting vs. Underfitting:

* **Overfitting:** The model learns the training data *too well* (including the noise), performing poorly on new data.
* **Underfitting:** The model is too simple to capture the underlying pattern of the data.

8. Evaluation Basics

    Accuracy: The percentage of correct predictions out of all predictions made.

    Precision: Of all the "positive" predictions the model made, how many were actually correct? (Focus on quality of positive predictions).

    Recall: Of all the actual positive cases, how many did the model correctly identify? (Focus on catching all positive cases).

    F1 score: The harmonic mean of Precision and Recall—a balanced metric when you need a single number to judge performance.

    Mean squared error (MSE): A common metric for Regression models; it measures the average squared difference between estimated values and the actual value. Lower is better.
