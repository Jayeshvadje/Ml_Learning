```python
from random import seed

import pandas as pd
import numpy as np

np.random.seed(42)
data = {
    'Test_Name': [f'TC_{i}' for i in range(1, 21)],
    'Status': np.random.choice(['Pass', 'Fail', 'Flaky'], 20, p=[0.7, 0.2, 0.1]),
    'Response_Time': np.random.randint(0,80,20),
    'Module': np.random.choice(['Login', 'Payments', 'Search'], 20)
}
df = pd.DataFrame(data)
```
