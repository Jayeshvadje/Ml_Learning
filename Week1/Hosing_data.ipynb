```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Load data
housing = fetch_california_housing(as_frame=True)
df = housing.frame
```

```python
df["HouseAge"]=df["HouseAge"].astype(int)
```

```python
df["MedHouseVal"] = df["MedHouseVal"]*10000
df["MedHouseVal"]=df["MedHouseVal"].astype(int)
df["Population"] =df["Population"].astype(int)
```

```python
df.head()
```

```python
sns.histplot(df,x="MedHouseVal",kde=True)
```

```python
fig = px.histogram(df, x="MedHouseVal",
                   title="Distribution of Median House Value (Target Variable)",
                   marginal="box", # Adds a box plot to show skew and outliers
                   nbins=50,
                   labels={'MedHouseVal': 'House Value ($100k units)'})

fig.show()
```

```python
df["MedHouseVal"].skew()
```

```python
subset = df[['MedInc', 'AveRooms', 'MedHouseVal']]
corr_matrix = subset.corr()
sns.heatmap(corr_matrix)
```

```python


import plotly.express as px

fig = px.density_heatmap(df, x="MedInc", y="AveRooms",
                         title="Density Heatmap of Income vs Rooms",
                         marginal_x="histogram",
                         marginal_y="histogram")
fig.show()
```
