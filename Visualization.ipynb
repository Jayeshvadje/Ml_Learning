```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("titanic.csv")
```

```python
sns.histplot(df['Age'], bins=20, color='red')
plt.show()
```

```python
sns.histplot(df['Age'],bins=20, color='red')
plt.title("Histogram of Age",fontsize=10,fontweight='bold')
plt.xlabel("Age")
plt.ylabel("Numebr of passengers")
plt.grid(True,linestyle='--',alpha=0.9)
plt.show()
plt.show()

```

```python
# Using 'hue' automatically generates a legend in Seaborn
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='viridis')

# 5. Legend Customization
plt.legend(title='Status', labels=['Perished', 'Survived'], loc='upper right')
```

```python
plt.figure(figsize=(12, 6))
sns.set_theme(style="darkgrid")

# Plotting
sns.boxplot(data=df, x='Pclass', y='Age', hue='Sex')

# Customization
plt.title('Age Distribution by Class and Gender', fontsize=15)
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.xticks([0, 1, 2], ['1st Class', '2nd Class', '3rd Class']) # Rename ticks
plt.grid(axis='y', alpha=0.3)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.set_theme(style="ticks")
sns.boxplot(data=df,x="Pclass",y="Fare",hue="Sex")
plt.title("Fare discribution by class",fontstyle="italic",fontsize=15)
plt.xlabel("passenger class")
plt.ylabel("Fare of passenger")
plt.xticks([0,1,2],["1st Class","2nd Class","3rd Class"])
plt.grid(axis="y",alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.set_theme(style="darkgrid")
plt.title("Performance Test Results")
```

```python
sns.countplot(data=df, x='Pclass', palette='Paired')
plt.legend(title='Test Result', loc='upper left')
plt.show()
```

```python
# Create a layout with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1 on the left
sns.histplot(df['Fare'], ax=axes[0])
axes[0].set_title("Distribution")

# Plot 2 on the right
sns.boxplot(y=df['Fare'], ax=axes[1])
axes[1].set_title("Outliers")
```
